```python
# set variables
comment_dict=[]
youtube = build(API_SERVICE_NAME, API_VERSION,developerKey=app_secret_key)


# define the function to get data from Youtube API 
def get_comments(youtube, video_id):


# get the information of video features with videos.list    
    results_v = youtube.videos().list(
        part="snippet,statistics",
        id=video_id,
    ).execute()

# get the information of comment with commentThreads.list  
    results_c = youtube.commentThreads().list(
        part="snippet",
        maxResults=100,
        videoId=video_id,
        textFormat="plainText"
    ).execute()
      
    
    name=[]
    language=[]
    vlike=[]
   
    comment=[]
    comment_count=[]
    cid=[]
    vid = [] 
    reply=[] 
    like = []
    dislike=[]
    publish=[] 
    author = []
    author_id=[]
    text=[]
    

# assign data to corresponding variables  
    for item in results_v["items"]:
        name = item["snippet"]["title"]
        vlike = item['statistics']['likeCount']
        dislike = item['statistics']['dislikeCount']
        
        comment_count= item['statistics']['commentCount']
          
    for item in results_c["items"]:
        cid = item['id']
        reply = item["snippet"]["totalReplyCount"]
        comment = item["snippet"]["topLevelComment"]
        like = comment["snippet"]["likeCount"]
        vid = comment["snippet"]["videoId"]
        publish = comment["snippet"]["publishedAt"]
        author = comment["snippet"]["authorDisplayName"]
        author_id = comment["snippet"]["authorChannelId"]['value']
        text = comment["snippet"]["textDisplay"]
       
    
        dict={'Comment_id':cid,'Comment':text,'Comment_reply':reply,
              'Comment_like':like,'Comment_published':publish, 
              'Video_id':vid,'Video_title':name,'Video_like':vlike,'Video_dislike':dislike,'Comment_count':comment_count,
              'Author':author,'Author_id':author_id 
              } 
        comment_dict.append(dict)
        global re
        
        
    

# dataframe dataset     
    re=pd.DataFrame(comment_dict)
    re =re.reindex(columns=['Author', 'Author_id','Comment','Comment_id','Comment_like','Comment_reply','Comment_published'
                            ,'Video_title','Video_id','Comment_count','Video_like','Video_dislike'])
    return  re 
   
```
